## Basic Usage

### Run pipelines

```python
from shared import RSSPipeline, APIPipeline

RSSPipeline(regions=["usa", "spain"], tickers=["AAPL", "MSFT", "PLUG", "ASML", "IBE.MC", "EBRO.MC"], max_age_hours=72).run(cleanup=True) 
APIPipeline(tickers=["AAPL", "MSFT", "PLUG", "ASML", "IBE.MC", "EBRO.MC"]).run(cleanup=True)
```

### Search

```python
from shared import SearchClient

client = SearchClient()

results = client.search(
    query="ASML",
    mode="hybrid",
    published_after="2025-12-07",
    keywords=["ASML"],
    keyword_boosts={"ASML": 2.0})

for result in results:
    print(f"Score: {result['score']}, ID: {result['id']}, Collection: {result['collection']}")
    print(f"Published date: {result["metadata"]["published_date"]}")
    print(f"Text: {result.get('text', 'N/A')}")
    print("---")

print(f"Found {len(results)} results")
```
